# 🚢 Titanic Survival Analysis

## 📋 Project Overview
The sinking of the Titanic on April 15, 1912, remains one of the most tragic maritime disasters in history. This project analyzes passenger data to understand what factors most influenced a person's chances of survival.

Using Python's pandas library, I explore patterns in the data and answer key questions about who lived, who died, and why.

## 🎯 Main Goal
**To identify the key factors that determined survival** by analyzing passenger characteristics such as gender, passenger class, age, family size, and fare.

## 📊 Key Questions
1. Did gender affect survival?
2. Did passenger class matter?
3. Did age play a role?
4. Did family size influence survival?
5. Did fare amount impact survival?

## 🔍 Finding 1: Gender and Survival

### Key Result
**Women had a significantly higher survival rate than men.**

- Female survival rate: **74.2%**
- Male survival rate: **18.9%**
- Difference: **55.3 percentage points**

### What This Means
Gender was the strongest predictor of survival. Women were almost **4 times more likely** to survive than men. This confirms historical accounts that the "women and children first" protocol was strictly followed during the evacuation.

### Interpretation
Out of 314 female passengers, 233 survived. Out of 577 male passengers, only 109 survived. The stark contrast indicates that being female dramatically increased a passenger's chances of survival.

## 🔍 Finding 2: Passenger Class and Survival

### Key Result
**Passengers in 1st class had the highest survival rate, while 3rd class passengers had the lowest.**

| Passenger Class | Survival Rate |
|-----------------|---------------|
| 1st Class       | **63.0%** |
| 2nd Class       | **47.3%** |
| 3rd Class       | **24.2%** |

### What This Means
Passenger class played a major role in survival. Wealthier passengers in 1st class were far more likely to survive than those in 3rd class. This likely reflects their proximity to lifeboats and priority access during the evacuation.

### Interpretation
- 1st class passengers: 63% survived
- 2nd class passengers: 47% survived
- 3rd class passengers: only 24% survived

The pattern is clear: **higher class = higher survival rate**.

## 🔍 Finding 3: Age and Survival

### Key Result
**Children had the highest survival rate, while seniors had the lowest.**

| Age Group | Survival Rate |
|-----------|---------------|
| Child (0-12)   | **58.0%** |
| Teen (13-19)   | **42.9%** |
| Adult (20-39)  | **38.3%** |
| Middle-aged (40-59) | **40.0%** |
| Senior (60+)   | **22.7%** |

### What This Means
Age clearly influenced survival. Children were prioritized during the evacuation, consistent with the "women and children first" protocol. Seniors had the lowest survival rate, possibly due to mobility issues or being less able to reach lifeboats.

### Interpretation
- Children (0-12): **58%** survival – highest of all groups
- Teenagers: **43%** survival
- Adults: **38%** survival
- Middle-aged: **40%** survival
- Seniors (60+): **23%** survival – lowest of all groups

**Key Pattern:** Survival decreases with age, with children being the clear priority and seniors being the most vulnerable.

## Finding 4: Fare and Survival

### Key Result
**Passengers who paid higher fares had significantly higher survival rates.**

| Fare Category | Survival Rate |
|---------------|---------------|
| Very Low | 32.4% |
| Low | 65.4% |
| Medium | 75.8% |
| High | 70.0% |

### What This Means
Higher fares were strongly associated with higher survival. Passengers in the lowest fare category had only a 32% survival rate, while those in medium and high categories had over 70% survival.

### Key Insight
The survival rate jumps by over 33 percentage points from Very Low to Low category. This suggests that even a small increase in ticket price dramatically improved survival chances.

### Why the High Category Drop?
The slight drop at the highest fare level (70.0% vs 75.8%) could be due to:
- Small sample size in the High category
- Some high fare passengers being families with multiple tickets
- Random variation in the data

**Overall Conclusion:** Fare was a significant factor in survival. Wealthier passengers had a clear advantage.

## 🔍 Finding 5: Embarkation Port and Survival

### Key Result
**Passengers who boarded at Cherbourg (C) had the highest survival rate.**

| Port | Survival Rate |
|------|---------------|
| Cherbourg (C) | **55.4%** |
| Queenstown (Q) | **39.0%** |
| Southampton (S) | **33.7%** |

### What This Means
Where a passenger boarded the Titanic significantly affected their survival chances. Those who boarded at Cherbourg had a survival rate over 20 percentage points higher than those who boarded at Southampton.

### Why Did Cherbourg Passengers Survive More?

**Possible Reasons:**

1. **Class Distribution** - Most 1st class passengers boarded at Cherbourg, while many 3rd class passengers boarded at Southampton

2. **Proximity to Lifeboats** - 1st class cabins were closer to lifeboats, giving them better access during evacuation

3. **Boarding Location** - Passengers from Cherbourg were more likely to be in 1st class, which had priority access to lifeboats

4. **Social Factors** - Wealthier passengers (who boarded at Cherbourg) were more likely to be prioritized

### Connection to Previous Findings

This finding is closely related to the **Class** analysis:
- Cherbourg had many 1st class passengers → High survival
- Southampton had many 3rd class passengers → Low survival

**Key Insight:** The port effect is likely driven by the mix of passenger classes, not the port itself.

### What This Means
Having **one** family member (sibling or spouse) on board significantly improved survival chances. This suggests:

1. **Support matters** - Having someone to help you during evacuation
2. **Small groups are efficient** - Easy to stay together and coordinate
3. **Women/children in small groups** - If you were with a woman or child, you were prioritized

### Why It Drops for Larger Groups

**2. Wealth Mattered**

### What This Tells Us

The Titanic evacuation was **not random** - it followed clear patterns:

1. **Women and children were prioritized** (consistent with maritime tradition)
2. **Wealthy passengers had advantages** (better access to lifeboats)
3. **Social hierarchy mattered** (class determined survival chances)

**The tragedy revealed both the best and worst of human behavior** - chivalry in prioritizing women and children, but also class inequality in survival outcomes.

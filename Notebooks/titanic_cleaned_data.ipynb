This notebook covers the **data cleaning and preprocessing** phase of the Titanic project. The goal was to take raw passenger data and transform it into a **clean, analysis-ready dataset**.

### Column Descriptions

| Column | Description |
|--------|-------------|
| `PassengerId` | Unique ID for each passenger |

| `Name` | Passenger name |
| `Sex` | Gender (male/female) |

| `SibSp` | # of siblings/spouses aboard |
| `Parch` | # of parents/children aboard |
| `Ticket` | Ticket number |
| `Fare` | Passenger fare |
| `Cabin` | Cabin number |
| `Embarked` | Port of embarkation (C=Cherbourg, Q=Queenstown, S=Southampton) |

## 🔍 Handling Missing Values

### Missing Values Before Cleaning

| Column | Missing Values | Action Taken |
|--------|---------------|--------------|
| Age | 177 | Filled with **median age** (28.0) |
| Embarked | 2 | Filled with **most common port** (S) |
| Cabin | 687 | **Dropped** column (too many missing) |

### Why These Choices?

| Column | Why This Method |
|--------|-----------------|
| **Age** | Median is robust to outliers (no extreme values affect it) |
| **Embarked** | Mode (most common value) is best for categorical data |
| **Cabin** | Over 75% missing → dropping is the safest option |

## 🔄 Converting Data Types

### Categorical to Numeric

To make the data usable for analysis and machine learning, categorical columns were converted to numeric.

| Column | Before | After |
|--------|--------|-------|
| **Sex** | male/female | 0 (male), 1 (female) |
| **Embarked** | C/Q/S | 0 (C), 1 (Q), 2 (S) |

### Why This Matters

- Machine learning models only understand **numbers**
- Converting text to numbers allows us to find **patterns**
- This is called **label encoding**

### Data Type Optimization

| Column | Original Type | New Type |
|--------|---------------|----------|
| Pclass | int64 | category |
| Sex | object | category |
| Embarked | object | category |

## ✨ Feature Engineering (Creating New Columns)

New columns were created to reveal patterns in the data.

### 1. AgeGroup

| Age Range | Group Name |
|-----------|------------|
| 0-12 | Child |
| 13-19 | Teen |
| 20-39 | Adult |
| 40-59 | Middle-aged |
| 60+ | Senior |

**Purpose:** Age as categories shows clearer survival patterns than continuous ages.

### 2. Family_Size

`Family_Size = SibSp + Parch`

**Purpose:** Shows total family members on board. Alone passengers vs those with family.

### 3. Fare_Category

| Fare Range | Category |
|------------|----------|
| $0-50 | Very Low |
| $50-100 | Low |
| $100-200 | Medium |
| $200+ | High |

**Purpose:** Fare as categories makes it easier to see patterns.
